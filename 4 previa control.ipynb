```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split 
from keras.utils import to_categorical

data = load_iris()
#print(data)
dataX, dataY = data['data'], data['target']

Xtrain, Xtest, Ytrain, Ytest  = train_test_split(dataX, dataY, test_size=0.2, shuffle=True, random_state=1)

Ytrain = to_categorical(Ytrain)
Ytest  = to_categorical(Ytest)

print(Ytrain)
print(Ytest)
```

Construya una NN para el dataset iris con 2 capas ocultas con 20 y 40 neuronas cada una y entrenela.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

NN = Sequential()
NN.add(Dense(20, input_shape=(4,), activation="relu" ))
NN.add(Dropout(0.5))
NN.add(Dense(40, activation="relu" ))
NN.add(Dense(3, activation="sigmoid")) # son solo 3 clases

NN.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["acc"]) 

NN.summary()
```

```python
# parámetros entrenables
4*20+20    +    20*40+40    +    40*4+4
```

```python
historia = NN.fit(Xtrain, Ytrain, epochs=1000, validation_data = (Xtest, Ytest))
```

```python
import matplotlib.pyplot as plt 

plt.plot(historia.history['loss'])
plt.plot(historia.history['val_loss'])
```

```python
plt.plot(historia.history['acc'])
plt.plot(historia.history['val_acc'])
```

```python
print(NN.predict([ [1, 2, 3, 4 ], [2,3,4,5]  ]) )
```

# Construir una NN para regresion


```python
from sklearn.datasets import load_diabetes

data = load_diabetes()
X,Y = data['data'], data['target'] # deberiamos dividirlo en train y test (no lo hicimos por tiempo)
```

```python
# construir una red neuronal para regresion con 3 capas ocultas de 10 20 y 30 neuronas 
model = Sequential()
model.add(Dense(10, input_shape=X[0].shape, activation="relu" )) # input y oculta 1
model.add(Dense(20, activation="relu")) # oculta 2
model.add(Dense(30, activation="relu")) # oculta 3
model.add(Dense(1))  # la salida tiene solo 1 neurona (porque busca predecir 1 numero), sin activación

model.compile(optimizer="adam", loss="mse", metrics=["mae"])

model.fit(X, Y, epochs=100)
```

```python
model.predict([[0.1,1,2,3,4,5,6,7,8,10]])
```
